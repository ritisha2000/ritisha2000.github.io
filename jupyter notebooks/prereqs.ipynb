```python
import pandas as pd
```

```python
course_list = pd.read_excel('data\Course_Section_Search_-_Central.xlsx')
course_list.columns
```

```python
course_list = course_list.drop_duplicates(subset='Course')
course_list.shape
```

```python
course_list['Description']
```

```python
import re

course_description = course_list[['Course', 'Description']].copy()
course_description.loc[:, 'reqs'] = course_description['Description'].str.extract(
    r'((prerequisite|corequisite)[\s\S]*)', 
    flags=re.IGNORECASE
)[0]
course_description
```

```python
course_description['reqs'][78]
```

```python
pattern = r'\b[A-Z]{4}\s\d{3}\sis\srecommended\.'

# Remove the sentence if present
course_description['reqs'] = course_description['reqs'].str.replace(pattern, '', regex=True)
course_description
```

```python
def extract_reqs(text, keyword):
    if isinstance(text, str):
        match = re.search(fr'{keyword}:.*?[.\]]', text)
        return match.group(0) if match else ''
    return ''

# Create 'prereqs' and 'coreqs' columns
course_description['prereqs'] = course_description['reqs'].apply(lambda x: extract_reqs(x, 'Prerequisite'))
course_description['coreqs'] = course_description['reqs'].apply(lambda x: extract_reqs(x, 'Corequisite'))
course_description
```

```python
course_description['prereq_courses'] = course_description['prereqs'].apply(
    lambda x: re.findall(r'[A-Z]{4}\s*\d{3}', str(x)) if isinstance(x, str) else []
)
course_description['coreq_courses'] = course_description['coreqs'].apply(
    lambda x: re.findall(r'[A-Z]{4}\s*\d{3}', str(x)) if isinstance(x, str) else []
)
course_description
```

```python
course_description.drop(columns=['prereqs', 'coreqs', "Description"], inplace=True)
course_description["prereq_courses"] = course_description["prereq_courses"].apply("; ".join)
course_description["coreq_courses"] = course_description["coreq_courses"].apply("; ".join)
```

```python
course_w_preqs = course_list.merge(course_description, on="Course", how="left")

course_w_preqs.rename(columns={"Course": "course_id"}, inplace=True)
course_w_preqs["course_id"] = course_w_preqs["course_id"].str.replace("_V", "")
course_w_preqs = course_w_preqs[course_w_preqs["Academic Level"] == "Undergraduate"]
course_w_preqs = course_w_preqs[[
    "course_id", "Course Subject", "Course Number", "Section Title", 'Eligbility Rules', 
    "Term", "Delivery Mode", "Description", "prereq_courses", "coreq_courses"
]].copy()


course_w_preqs
```

```python
## After completing manual review
course_final = pd.read_excel("course_w_preqs.xlsx")
course_final.sample(3)
```

```python
from json import loads, dumps
import json

result = course_final.to_json(orient="index")
nodes = loads(result)

course_data = {}
course_data["nodes"] = list(map(dict, nodes.values()))
```

```python
def format_reqs(df, colname):
    temp = df[["course_id", colname]].copy()
    temp[colname] = temp[colname].str.split('; ')
    temp = temp.explode(colname).rename(columns={
        "course_id": "source",
        colname: "target"
    }).dropna().copy()
    temp["link_type"] = colname.split('_')[0]
    return temp

prereqs = format_reqs(course_final, "prereq_courses")
coreqs = format_reqs(course_final, "coreq_courses")

result = pd.concat([prereqs, coreqs], ignore_index=True).to_json(orient="index")
links = loads(result)

course_data["links"] = list(map(dict, links.values()))

```

```python
with open("course_data.json", 'w') as f:
    json.dump(course_data, f)
```
